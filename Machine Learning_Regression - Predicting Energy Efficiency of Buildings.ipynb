```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
energy_data = pd.read_csv(r"C:\Users\PHOSI\Downloads\energydata_complete.csv", error_bad_lines=False)
energy_data.head()
```

```python
energy_data.isna().sum()
```

```python
energy_data.describe()
```

```python
ml_data = energy_data.drop(columns=["date"])
ml_data
```

```python
corr_df = ml_data.corr()
corr_df[corr_df["T2"]>=0.7]
```

```python
# LINEAR MODEL T2 AND T6
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalized_df = pd.DataFrame(scaler.fit_transform(ml_data), columns=ml_data.columns)

predictor = normalized_df.drop(columns=["T1","T4", "T3",'T6', "T5", "T_out"])
response = normalized_df['T6']


from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(predictor, response, test_size=0.3, random_state=2)

from sklearn.linear_model import LinearRegression, Ridge, Lasso
model = LinearRegression()

model.fit(x_train, y_train)
predicted_val = model.predict(x_test)
```

```python
# R-squared value
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
rsquared = r2_score(y_test, predicted_val)
round(rsquared, 2)
```

```python
# Mean Absolute Error
mae = mean_absolute_error(y_test, predicted_val)
round(mae, 2)
```

```python
# Residual Sum of Squares
rss= np.sum(np.square(y_test - predicted_val)) 
round(rss, 3)
```

```python
# Root Mean Square Error
rmse=np.sqrt(mean_squared_error(y_test, predicted_val))
round(rmse, 3)
```

```python
# RIDGE REGRESSION
ridge = Ridge(alpha=0.001)
ridge.fit(x_train, y_train)
ridge.coef_
```

```python
ridge_val = ridge.predict(x_test)
rmse=np.sqrt(mean_squared_error(y_test, ridge_val))
round(rmse, 3)
```

```python
# LASSO REGRESSION
lasso = Lasso(alpha=0.001)
lasso.fit(x_train, y_train)
lasso.coef_
```

```python
lasso_val = lasso.predict(x_test)
rmse=np.sqrt(mean_squared_error(y_test, lasso_val))
round(rmse, 3)
```
