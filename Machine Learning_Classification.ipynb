```python
df.drop(columns=['stab'], inplace=True)
```

```python
y=pd.get_dummies(df['stabf'] )
y=y['stable']

df.drop(columns=['stabf'], inplace=True)
x = df
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
norm_x =  sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(norm_x, y, random_state=1, test_size=0.2)

import imblearn
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state= 1)
x_train_bal, y_bal = smote.fit_resample(x_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score

rfc = RandomForestClassifier(n_estimators=300, random_state=1)
rfc.fit(x_train_bal, y_bal)

y_rfc= rfc.predict(x_test)
accuracy_score(y_rfc, y_test)
# print(metrics.classification_report(y_rfc, y_test))
```

```python
from sklearn.ensemble import ExtraTreesClassifier
etc = ExtraTreesClassifier(n_estimators=300,
                           min_samples_split=9,
                           min_samples_leaf=2,
                           max_features='auto' ,random_state = 1)
etc.fit(x_train_bal, y_bal)

y_etc= etc.predict(x_test)
feature_importance = etc.feature_importances_
```

```python
import xgboost as xgb
xgb=xgb.XGBClassifier(n_estimators=100, use_label_encoder=False,random_state=1, eval_metric='error')
xgb.fit(x_train_bal, y_bal)

y_xgb= xgb.predict(x_test)

# print(metrics.classification_report(y_xgb, y_test))
accuracy_score(y_xgb, y_test)
```

```python
from lightgbm import LGBMClassifier
lgbm = LGBMClassifier(random_state=1, n_estimators=100)
lgbm.fit(x_train_bal, y_bal)

y_lgbm = lgbm.predict(x_test)

# print(metrics.classification_report(y_lgbm, y_test))
accuracy_score(y_lgbm, y_test)



```

```python
from sklearn.model_selection import RandomizedSearchCV

n_estimators = [50, 100, 300, 500, 1000]
min_samples_split = [2, 3, 5, 7, 9]
min_samples_leaf = [1, 2, 4, 6, 8]
max_features = ['auto', 'sqrt', 'log2', None]

hyperparameter_grid = {'n_estimators': n_estimators,
                       'min_samples_leaf': min_samples_leaf,
                       'min_samples_split': min_samples_split,
                       'max_features': max_features}
rscv = RandomizedSearchCV(etc, hyperparameter_grid, random_state=1)
```
